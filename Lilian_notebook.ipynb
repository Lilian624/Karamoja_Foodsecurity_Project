```python
# Renaming columns
# Standardize lower case & upper case
subcounty_info.columns = subcounty_info.columns.str.strip().str.lower().str.replace('_', ' ')
district_info.columns = district_info.columns.str.strip().str.lower().str.replace('_', ' ')

#Print the first 5 rows to check if the cleaning is done
print('Here are the 5 first rows of subcounty info data')
subcounty_info.head(5)
```

```python
print('Here are the 5 first rows of District info data')
district_info.head(5)
```

Renaming columns that have similar data that is Name in the subcounty info data set and the subcounty name in the district info data set
After renaming them next step would be to merge the two data sets

```python
#Renaming the Name column in the Sub count Info Data Set
subcounty_info.rename(columns={'name': 'subcounty_name'}, inplace=True)
#print the first 5 rows to check the changes
subcounty_info.head(5)
```

```python
print(subcounty_info)
```

```python
district_info.head(5)
```

```python
print(subcounty_info.columns)
print(district_info.columns)
```

```python

district_info = district_info.rename(columns={'subcounty name': 'subcounty_name'})
```

```python
# Merge the two datasets , reason is now the sub county info data set has a subcounty name column and so is the District info data set

merged_dataset = pd.merge(subcounty_info, district_info, on='subcounty_name', how='outer')
#print the merged data set both its information and the first 5 rows
print(merged_dataset.info())
print(merged_dataset.head())
```

```python
print("This is the Summary of the Merged Dataset")
print(merged_dataset.describe())
```
